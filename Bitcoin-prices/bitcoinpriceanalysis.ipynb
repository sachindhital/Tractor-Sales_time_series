```python
# loading the data and setting the index
df = pd.read_csv("BTC-USD.csv",index_col = "Date",parse_dates= True)
```

```python
df.head()
```

```python
df.index
```

```python
# resampling to monthly frequency and calculate the mean closing price
df.resample('W').mean()
```

# Exploring Data

```python
# 7 - day rolling average of the closing price 
df['7_day_rolling']=df['Close'].rolling(window = 7).mean()
```

```python
df[['Close','7_day_rolling']].loc['2023'].plot()
plt.show()
```

```python
# find out the highest average month 
df.resample('M').mean()['Close'].idxmax()
```

```python
#calculate Daily Returns
df['daily_returns'] = df['Close'].pct_change()*100
```

```python
# Days with more than 10% change in closing price 
df[abs(df['daily_returns']>10) ].head()
```

# Data Visualization

```python
# Daily closing priceplot
df['Close'].plot(title='Daily Closing Price')
plt.show()
```

```python
# Plot the yearly Volume 
df.resample('YE').sum()['Volume'].plot()
```

```python
#plotting closing price and 30- day rolling volume
df['30_day_rolling']= df['Volume'].rolling(window=30).mean()
df['30_day_rolling'].plot(legend=True)
ax  = df['Close'].plot(secondary_y=True, legend=True)
ax.set_ylabel('Volume')
plt.show()
```

```python
# correlation between the 30-dat rolling volume and closing price
df[['Close','30_day_rolling']].corr()
```

Data Manipulation

```python
# Identify missing values
df.isnull().sum()

#fill missing values 
df['30_day_rolling'].fillna(method = 'bfill',inplace=True)

#interpolate missing 7 day rolling 
df['7_day_rolling'].interpolate(inplace=True)       
```

```python
# Identify missing values
df.isnull().sum()
```

```python
df['year']= df.index.year
df['month'] = df.index.month 
df['day'] = df.index.day 
df['weekday'] = df.index.day_name()
df['weekday_numeric'] = df.index.weekday

df['is_weekend'] = df.index.weekday>4
df.head()
```

```python
# Feature Engineering - lagged values 
df['closed_lag1']=df['Close'].shift(1)
df['closed_lag2']=df['Close'].shift(2)
```

Seasonal Decomposition Identifies the trend, seasonality, and error term

Seasonality

```python
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
```

```python
#plotting the monthly seasonality 
month_plot(df['Close'].resample('ME').mean(),ylabel='Closing')
plt.show()
```

```python
#plotting the monthly seasonality 
quarter_plot(df['Close'].resample('Q').mean(),ylabel='Closing')
plt.show()
```

```python
# load new data 
df_choco = pd.read_csv('C:\\Projects\\Time_series\\Bitcoin-prices\\Introduction to Time Series Forecasting\\choco_monthly_revenue.csv',index_col=0,parse_dates=True)
df_choco.head()
```

```python
#plotting the monthly seasonality 
month_plot(df_choco['revenue'],ylabel='Closing')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#seasnoal decomposition plots for bitcoin data
decomposition = seasonal_decompose(df['Close'],
                   model= 'mul', 
                   period=365)

fig = decomposition.plot()
fig.set_size_inches(10,8)
plt.show()
```

```python
#seasnoal decomposition plots for chocolate data
decomposition = seasonal_decompose(df_choco['revenue'],
                   model= 'mul', 
                   period=12)

fig = decomposition.plot()
fig.set_size_inches(10,8)
plt.show()
```

Seasonality:
* 24 for hourly 
* 6 or 365 for daily but 7 is preferred 
* for modeling 
* 52 for weekly 
* 12 for monthly 
* 4 for quarterly 
* 5 for weekdays 

Auto- correlation tells us if there is information in the past

The process is to correlate the time series with its lagges values

```python
from statsmodels.graphics.tsaplots import plot_acf 
```

```python
#plot the autocorrelation (ACF) 
fig, ax = plt.subplots(figsize = (10,6))
plot_acf(df['Close'],lags=100, ax=ax)
plt.show()
```

```python
#plot the autocorrelation (ACF) 
fig, ax = plt.subplots(figsize = (10,6))
plot_acf(df_choco['revenue'],lags=30, ax=ax)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf 
```

# Partial Auto correlation

```python
#plot the partial autocorrelation (PACF) 
fig, ax = plt.subplots(figsize = (10,6))
plot_pacf(df['Close'],lags=100, ax=ax)
plt.show()
```

```python
#plot the partial autocorrelation (PACF) 
fig, ax = plt.subplots(figsize = (10,6))
plot_pacf(df_choco['revenue'],lags=30, ax=ax)
plt.show()
```
